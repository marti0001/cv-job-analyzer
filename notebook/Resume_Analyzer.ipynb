## asystent analizy stanowiska 

Na podstawie opisu stanowiska pracy wyciagne najwazniejsze informacje w uporzƒÖdowany spos√≥b 

```python
#%%writefile job_description_prompt.py

from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')

if "OPENAI_API_KEY" not in os.environ:
    raise ValueError("Brak klucza API w zmiennych ≈õrodowiskowych")

openai = OpenAI(api_key=openai_api_key)

def job_data_prompt(resume_data):
    system_prompt= '''Jako asystent, Twoim zadaniem jest przeanalizowanie opisu stanowiska pracy i wypisanie jego kluczowych element√≥w w strukturalny spos√≥b. 
    Dla ka≈ºdego opisu wykonaj analizƒô i zwr√≥ƒá odpowied≈∫ w nastƒôpujƒÖcym formacie (bez u≈ºycia markdown):
    Stanowisko: [nazwa stanowiska]
    Firma: [nazwa firmy]
    Lokalizacja: [miasto, region, kraj ‚Äì je≈õli dostƒôpne]
    Tryb pracy: [zdalny / hybrydowy / stacjonarny / opcjonalnie]
    Typ zatrudnienia: [pe≈Çen etat / czƒô≈õƒá etatu / umowa zlecenie / B2B / sta≈º / praktyki / inny]
    Poziom stanowiska: [junior / mid / senior / kierownicze / specjalista]
    Opis obowiƒÖzk√≥w:
    - [punkt 1]
    - [punkt 2]
    - [punkt 3 itd.]
    
    Wymagania kandydata:
    - [punkt 1]
    - [punkt 2]
    - [punkt 3 itd.]
    
    Zalety oferty:
    - [punkt 1]
    - [punkt 2]
    - [punkt 3 itd.]
    
    Link do oferty: [je≈õli dostƒôpny]
    
    Uwagi: [opcjonalnie ‚Äì np. czy wymagana znajomo≈õƒá jƒôzyka, czy sƒÖ benefity, inne szczeg√≥≈Çy].
    Zwr√≥ƒá uwagƒô na priorytety w opisie (np. ‚Äûkonieczne‚Äù, ‚Äûopcjonalne‚Äù, ‚Äûwysoko cenione‚Äù).
    Dzia≈Çaj dok≈Çadnie i nie pomijaj istotnych szczeg√≥≈Ç√≥w. Je≈õli jaka≈õ informacja nie jest dostƒôpna w tek≈õcie, pozostaw pole puste lub napisz ‚Äûnie okre≈õlono‚Äù. 
    Przetwarzaj tekst tak, jakby≈õ przygotowywa≈Ç dane do bazy ofert pracy. U≈ºywaj jƒôzyka polskiego. 
    Nie dodawaj ≈ºadnych komentarzy poza formatem wynikowym.
    '''    
    prompt = [
        {'role': 'system', 'content': system_prompt}
    ]
    
    user_prompt = resume_data

    prompt.append({'role':'user', 'content': user_prompt})
    
    response = openai.chat.completions.create(
        model= "gpt-4o-mini", 
        messages=prompt,
        temperature= 0.0,
    )
    
    data = response.choices[0].message.content

    return data
```

## analiza cv 

na podstawie przes≈Çanego cv zidentyfikuje umiejetnosci kandydata do≈õwiadczenie zawodowe, edukacjƒô, jƒôzyki obce i inne istotne informacje. po czy por√≥wnam je z wymaganiami na stanowisko.
Dla ka≈ºdego elementu z opisu stanowiska sprawd≈∫, czy wystƒôpuje w CV:
Je≈õli tak ‚Üí przyznaj punkty za trafno≈õƒá i jako≈õƒá potwierdzenia (np. d≈Çugo≈õƒá do≈õwiadczenia, konkretna rola).
Je≈õli nie ‚Üí odejmij punkty lub obni≈º wagƒô.
WyciƒÖgnij wnioski dotyczƒÖce brak√≥w lub nadmiar√≥w kwalifikacji wzglƒôdem wymaga≈Ñ.
Podaj kr√≥tkie uzasadnienie oceny, wskazujƒÖc na g≈Ç√≥wne obszary pokrycia siƒô oraz braki wzglƒôdem oczekiwa≈Ñ pracodawcy.
Przelicz wynik na skalƒô od 1 do 100

```python
#%%writefile resume_analyzer.py

from PyPDF2 import PdfReader
from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')

if "OPENAI_API_KEY" not in os.environ:
    raise ValueError("Brak klucza API w zmiennych ≈õrodowiskowych")

openai = OpenAI(api_key=openai_api_key)


def extract_text_from_pdf(pdf_path):
    with open(pdf_path, "rb") as file:
        reader = PdfReader(file)
        text = ''
        for page in reader.pages:
            text += page.extract_text()
        return text

def cv_prompt(cv_parser):
    
    system_prompt = """Jeste≈õ specjalistƒÖ ds. rekrutacji i analizy danych. Twoim zadaniem jest dok≈Çadnie przeanalizowaƒá przes≈Çane CV i wyodrƒôbniƒá z niego nastƒôpujƒÖce informacje w uporzƒÖdkowanej formie:
    - Umiejƒôtno≈õci:
    Techniczne (np. jƒôzyki programowania, narzƒôdzia, technologie)
    Miƒôkkie (np. komunikacja, praca w zespole)
    - Do≈õwiadczenie zawodowe:
    Dla ka≈ºdej pozycji:
    Nazwa firmy
    Rola / stanowisko
    Okres zatrudnienia
    G≈Ç√≥wne obowiƒÖzki i osiƒÖgniƒôcia
    Kluczowe umiejƒôtno≈õci zdobyte
    - Wykszta≈Çcenie:
    Kierunek studi√≥w
    Poziom (licencjat, magister itp.)
    Uczelnia
    Rok uko≈Ñczenia
    - Jƒôzyki obce:
    Jƒôzyk i poziom bieg≈Ço≈õci (np. B2, C1)
    Certyfikaty (je≈õli dostƒôpne)
    - Inne istotne informacje:
    Certyfikaty bran≈ºowe
    Kursy dodatkowe
    Projekty w≈Çasne lub open source
    Wolontariat / dzia≈Çalno≈õƒá spo≈Çeczna
    
    Zwr√≥ƒá wynik analizy w formacie JSON, oto przyk≈Çad:
    {
      "umiejetnosci": {
        "techniczne": ["Java", "Git", ...],
        "miekke": ["Komunikacja", "Praca w zespole", ...]
      },
      "doswiadczenie": [
        {
          "firma": "ABC Sp. z o.o.",
          "stanowisko": "Programista Java",
          "okres": "2020‚Äì2023",
          "obowiazki": ["Tworzenie aplikacji backendowych", "Udzia≈Ç w code review", ...],
          "osiagniecia": ["Zoptymalizowanie dzia≈Çania serwisu o 30%", ...]
        }
      ],
      "edukacja": [
        {
          "uczelnia": "Uniwersytet Jagiello≈Ñski",
          "kierunek": "Informatyka",
          "poziom": "Magister",
          "rok_ukonczenia": 2020
        }
      ],
      "jezyki_obce": [
        {
          "jezyk": "angielski",
          "poziom": "B2"
        }
      ],
      "inne_informacje": {
        "certyfikaty": ["Oracle Certified Java Programmer", ...],
        "projekty": ["Aplikacja do zarzƒÖdzania zadaniami ‚Äì GitHub", ...],
        "wolontariat": []
      }
    }
    """

    prompt = [
        {'role': 'system', 'content': system_prompt}
    ]
    
    user_prompt = cv_parser

    prompt.append({'role':'user', 'content': user_prompt})
    
    response = openai.chat.completions.create(
        model= "gpt-4o-mini", 
        messages=prompt,
        temperature= 0.0,
    )
    
    cv_description = response.choices[0].message.content

    return cv_description
```

## por√≥wnanie 

```python
#%%writefile job_matcher.py

from openai import OpenAI
from dotenv import load_dotenv
import os

load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')

if "OPENAI_API_KEY" not in os.environ:
    raise ValueError("Brak klucza API w zmiennych ≈õrodowiskowych")

openai = OpenAI(api_key=openai_api_key)


def cv_matcher(cv_json, job_description):
    
    system_prompt = """Jeste≈õ specjalistƒÖ ds. rekrutacji i doradcƒÖ karier zawodowych. Twoim zadaniem jest wykonanie **systematycznej analizy dopasowania kandydata do stanowiska** na podstawie:  
    1. **JSON z danymi CV kandydata** (zawiera do≈õwiadczenie, umiejƒôtno≈õci, edukacjƒô itp.),  
    2. **Opisu stanowiska pracy** dostarczonego przez u≈ºytkownika (wymagania, kompetencje, oczekiwania).  
    
    ---
    
    ### üîç **KROK PO KROKU: ANALIZA DOPASOWANIA**
    
    #### 1. **Ocena do≈õwiadczenia zawodowego**  
    - Sprawd≈∫, czy stanowiska w CV odpowiadajƒÖ **nazwie, sektorowi lub kluczowym obowiƒÖzkom** w opisie stanowiska.  
    - Oce≈Ñ **czas trwania do≈õwiadczenia** (np. "minimum 2 lata" ‚Üí sprawd≈∫ sumƒô lat w CV).  
    - Przeanalizuj opisy stanowisk w CV:  
      - Czy sƒÖ konkretne dzia≈Çania i efekty? (np. "Zwiƒôkszy≈Çem satysfakcjƒô klient√≥w o 20%" ‚Üí +2 pkt)  
      - Czy brakuje szczeg√≥≈Ç√≥w? (np. "Obs≈Çuga klienta" bez kontekstu ‚Üí 0 pkt)  
    
    #### 2. **Analiza umiejƒôtno≈õci (technicznych i miƒôkkich)**  
    - Por√≥wnaj umiejƒôtno≈õci z CV z wymaganiami w opisie stanowiska.  
    - Przyznaj punkty tylko za **potwierdzone umiejƒôtno≈õci** (np. "Obs≈Çuga CRM" ‚Üí szukaj w opisach stanowisk lub projektach).  
    - Brak umiejƒôtno≈õci krytycznych dla stanowiska (np. "Bieg≈Ço≈õƒá w Excelu" dla analityka danych) ‚Üí -5 pkt.  
    
    #### 3. **Edukacja i certyfikaty**  
    - Sprawd≈∫ zgodno≈õƒá kierunku studi√≥w z bran≈ºƒÖ lub wymaganiami stanowiska.  
    - Czy kandydat ma certyfikaty wymagane lub po≈ºƒÖdane? (np. Scrum Master, PMP).
    - czy kandydat zna podane technologie? (np. python, pandas)
    
    #### 4. **Jƒôzyki obce**  
    - Por√≥wnaj poziom jƒôzyka w CV z wymaganiami (np. "Angielski B2" vs "Angielski C1" ‚Üí -2 pkt).  
    
    #### 5. **Luki i ryzyka**  
    - Zidentyfikuj braki w danych (np. przerwy zawodowe, brak opis√≥w stanowisk).  
    - Uwzglƒôdnij niejednoznaczne informacje (np. "Do≈õwiadczenie w IT" bez konkretnych technologii).  
    
    ---
    
    ### üìä **SYSTEM PUNKTOWY (0‚Äì100)**  
    1. **Maksymalna liczba punkt√≥w** dla ka≈ºdego obszaru:  
       - Do≈õwiadczenie: 30 pkt  
       - Umiejƒôtno≈õci: 30 pkt (techniczne: 20 pkt, miƒôkkie: 10 pkt)  
       - Edukacja: 10 pkt  
       - Jƒôzyki: 10 pkt  
       - Dodatkowe atuty (np. certyfikaty, do≈õwiadczenie miƒôdzynarodowe): max 20 pkt  
    
    2. **Zasady przyznawania punkt√≥w**:  
       - Pe≈Çne spe≈Çnienie wymaga≈Ñ ‚Üí 100% punkt√≥w za dany obszar.  
       - Czƒô≈õciowe spe≈Çnienie ‚Üí punkty proporcjonalne do stopnia dopasowania.  
       - Brak krytycznych wymaga≈Ñ ‚Üí odejmij punkty (np. brak B2 angielskiego ‚Üí -5 pkt).  
    
    3. **Skala ko≈Ñcowa**:  
       - 90‚Äì100: Idealne dopasowanie  
       - 70‚Äì89: Wysokie dopasowanie  
       - 50‚Äì69: ≈örednie dopasowanie  
       - 0‚Äì49: Niskie dopasowanie  
    
    ---
    
    ### üìù **STRUKTURA WYNIKU** 
    1. **Firma:" np. Microsoft" , Stanowisko: "np. MLOps" **
    2. **Wynik: Np.78/100.**
    3. **Podsumowanie jako≈õciowe**:  
       - Obszary pokrycia siƒô z wymaganiami pracodawcy  
       - G≈Ç√≥wne braki wzglƒôdem oczekiwa≈Ñ  
       - Nadmiary kwalifikacji (opcjonalnie)  
    
    4. **Szczeg√≥≈Çowa ocena punktowa** dla ka≈ºdego obszaru (np. "Do≈õwiadczenie: 25/30- brak doswiaczenia w danym obszarze", "braki w technologii").  
    
    5. **Ocena ko≈Ñcowa (0‚Äì100)** + **uzasadnienie**:  
       - Np.: "Wynik: 78/100. Silne do≈õwiadczenie w obs≈Çudze klienta (28/30), ale brak do≈õwiadczenia w IT (-5 pkt)".  
    
    6. **Sugestie rozwijajƒÖce**:  
       - Co kandydat mo≈ºe dodaƒá/rozwijaƒá (np. "Rozszerzenie opisu projekt√≥w w CV o konkretne wyniki", "Uzupe≈Çnienie certyfikatu B2 angielskiego",
        praca nad projektami z danego dzia≈Çu"").  
    
    ---
    
    ### üß© **DODATKOWE WSKAZ√ìWKI**  
    - **Nie domy≈õlaj siƒô informacji** brakujƒÖcych w CV ‚Äì zaznacz je jako "Brak danych".  
    - **Unikaj subiektywnych ocen** (np. "WyglƒÖda kompetentnie") ‚Äì opieraj siƒô tylko na faktach z CV i opisie stanowiska.  
    - **U≈ºywaj jƒôzyka polskiego**, klarownego i profesjonalnego.  
    - **Nie stosuj formatowania Markdown** ‚Äì tylko tekst prosty.  
    """


    prompt = [
        {'role': 'system', 'content': system_prompt}
    ]
    
    prompt.append({"role": "user", "content": "CV kandydata:\n\n" + cv_json}),  # Pierwszy input (np. dane CV)
    prompt.append({"role": "user", "content": "Opis stanowiska:\n\n" + job_description})

    response = openai.chat.completions.create(
        model= "gpt-4o-mini", 
        messages=prompt,
        temperature= 0.0,
        stream=True
    )

    cv_description = ""
    for chunk in response:
        cv_description += chunk.choices[0].delta.content or ''
        yield cv_description
```

## interface

```python
%%writefile main.py

import gradio as gr
from PyPDF2 import PdfReader
from openai import OpenAI
from dotenv import load_dotenv
import os

from job_matcher import cv_matcher
from resume_analyzer import cv_prompt, extract_text_from_pdf
from job_description_prompt import job_data_prompt

load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')

if "OPENAI_API_KEY" not in os.environ:
    raise ValueError("Brak klucza API w zmiennych ≈õrodowiskowych")

openai = OpenAI(api_key=openai_api_key)


def summarize(pdf_file, job_description):
    # Process job description
    processed_job = job_data_prompt(job_description)
    
    # Extract and process CV text
    cv_text = extract_text_from_pdf(pdf_file)
    processed_cv = cv_prompt(cv_text)
    
    # Get matching results
    result = cv_matcher(processed_cv, processed_job)
    yield from result
    
view = gr.Interface(
    fn=summarize,
    inputs=[
        gr.File(label="Upload CV (PDF)", file_types=[".pdf"]),
        gr.Textbox(label="Job Description", lines=5, placeholder="Paste job description here...")
    ],
    outputs=gr.Textbox(label="Matching Results"),
    title="CV Matching System",
    description="Upload a CV (PDF) and paste a job description to see matching results",
    flagging_mode="never"

)
if __name__ == "__main__":
    view.launch()
```
